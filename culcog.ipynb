```python
import random
import time
import ipywidgets as widgets
from IPython.display import display
import pandas as pd
from google.colab import files  # Colab에서 파일 다운로드를 위한 모듈

# 단어 배열 정의
questions = ["성장", "성공", "기회", "비전", "성취"]
tools = ["공부", "교양 영상 시청", "여윳돈 저축", "자격증 준비", "학회"]
experiences = ["외출", "오락 영상 시청", "여윳돈 소비", "휴식", "취미 동아리"]
positives = ["효율적", "도움이 되는", "가치 있는", "중요한", "필요한"]
negatives = ["비효율적", "쓸모없는", "가치없는", "중요하지 않은", "필요없는"]

# 결과 저장용 리스트
results = []

# 실험 참가자의 닉네임을 입력받는 위젯
nickname_input = widgets.Text(
    description="닉네임:",
    placeholder="이름을 입력하세요"
)

def start_experiment(b):
    nickname = nickname_input.value
    if not nickname:
        display(widgets.HTML(value="<strong>닉네임을 입력해주세요!</strong>"))
        return

    display(widgets.HTML(value=f"<h2>{nickname}님, 실험을 시작합니다!</h2>"))
    display(widgets.HTML(value="<p>총 18번의 테스트가 진행됩니다. 각 테스트에서 옵션을 빠르게 선택하세요.</p>"))
    time.sleep(2)
    ask_question(1, nickname)

# 닉네임 입력 후 버튼 클릭 시 실행
start_button = widgets.Button(description="실험 시작")
start_button.on_click(start_experiment)

# 닉네임 입력 필드와 버튼 표시
display(nickname_input)
display(start_button)

# 질문과 옵션 생성 함수
def generate_test(test_number):
    question = random.choice(questions)

    if test_number in [1, 2]:  # 도구 vs 경험
        left = random.choice(tools)
        right = random.choice(experiences)
    elif test_number in [11, 12]:  # 경험 vs 도구
        left = random.choice(experiences)
        right = random.choice(tools)
    elif test_number in [3, 4]:  # 긍정 vs 부정
        left = random.choice(positives)
        right = random.choice(negatives)
    elif test_number in [5, 6, 7, 8, 9, 10]:  # 도구+긍정 vs 경험+부정
        left = f"{random.choice(tools)} & {random.choice(positives)}"
        right = f"{random.choice(experiences)} & {random.choice(negatives)}"
    else:  # 경험+긍정 vs 도구+부정 (Tests 13-18)
        left = f"{random.choice(experiences)} & {random.choice(positives)}"
        right = f"{random.choice(tools)} & {random.choice(negatives)}"

    return question, left, right

# 실험을 위한 버튼 생성 및 반응 시간 측정 함수
def ask_question(test_number, nickname):
    question, left_option, right_option = generate_test(test_number)

    display(widgets.HTML(value=f"<h3>Test {test_number}</h3>"))
    display(widgets.HTML(value=f"<strong>질문:</strong> {question}"))
    display(widgets.HTML(value=f"1) {left_option}  <--->  2) {right_option}"))

    start_time = time.time()

    # 버튼 클릭 시 반응 시간 기록
    button_1 = widgets.Button(description=f"1) {left_option}")
    button_2 = widgets.Button(description=f"2) {right_option}")

    def on_button_click(choice):
        reaction_time = time.time() - start_time
        selected_option = left_option if choice == 1 else right_option
        results.append([nickname, test_number, question, selected_option, reaction_time])

        # 결과 출력
        display(widgets.HTML(value=f"<strong>선택한 옵션:</strong> {selected_option}"))
        display(widgets.HTML(value=f"<strong>반응 시간:</strong> {reaction_time:.3f}초"))

        # 다음 질문으로 넘어가기
        if test_number < 18:
            ask_question(test_number + 1, nickname)
        else:
            display(widgets.HTML(value="<h3>실험이 종료되었습니다. 결과를 출력합니다.</h3>"))
            save_results_to_excel(nickname)

    # 버튼 클릭 이벤트 연결
    button_1.on_click(lambda x: on_button_click(1))
    button_2.on_click(lambda x: on_button_click(2))

    # 버튼 표시
    display(button_1)
    display(button_2)

# 실험 결과를 엑셀 파일로 저장하고 다운로드 링크 제공
def save_results_to_excel(nickname):
    df = pd.DataFrame(results, columns=["닉네임", "테스트 번호", "질문", "선택한 옵션", "반응 시간"])
    file_name = f"reaction_times_{nickname}.xlsx"
    df.to_excel(file_name, index=False, engine='openpyxl')

    # Colab에서 파일 다운로드 링크 생성
    files.download(file_name)
    display(widgets.HTML(value=f"<h3>결과가 '{file_name}' 파일로 저장되었습니다. 파일을 다운로드하세요.</h3>"))
```
